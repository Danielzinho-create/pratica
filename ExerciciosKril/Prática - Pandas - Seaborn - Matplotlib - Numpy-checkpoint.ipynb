Test the Law Of Large Numbers for N random normally distributed numbers with
mean = 0, stdev = 1:

Create a Python script that will count how many of these numbers fall between -1
and 1 and divide by the total quantity of N

You know that E(X) = 68.2%

Check that Mean(XN) -> E(X) as you rerun your script while increasing N

```python
from numpy.random import randn #importa a função randn da biblioteca numpy
```

```python
nContador = 0 # Cria um contador
qtdRandom = 100 #Determina quantos números serão gerados randomicamente (Observe que a cada 0 adicionado, a porcentagem fica menos variante e mais no entorno de 68,2%)
for i in randn(qtdRandom): #cria um loop para avaliar se os números gerados atendem a condição abaixo
    if i < 1 and i > -1:
        nContador += 1 #Adiciona +1 ao contador caso a condição seja cumprida
    else:
        None 
prcnt = (nContador*100)/qtdRandom #Transforma em porcentagem a quantidade de números que atendem a condição
print(f"{prcnt}% dos números gerados estão entre -1 e 1")
```

Scenario: You are a Data Scientist working for a consulting firm. One of your colleagues from the Auditing department has asked you to help them assess the financial statement of organisation X.

You have been supplied with two lists of data: monthly revenue and monthly expenses for the financial year in question. Your task is to calculate the following financial metrics:
- profit for each month
- profit after tax for each month (the tax rate is 30%)
- profit margin for each month - equals to profit after tax divided by revenue
- good months - where the profit after tax was greater than the mean for the year
- bad months - where the profit after tax was less than the mean for the year
- the best month - where the profit after tax was max for the year
- the worst month - where the profit after tax was min for the year

All results need to be presented as lists.

Results for dollar values need to be calculated with 0.01 precision, but need to be presented in Units of $1,000 (i.e. 1k) with no decimal points.

Results for the profit margin ratio need to be presented in units of % with no decimal points.

Note: Your colleague has warned you that it is okay for tax for any given month to be negative (in accounting terms, negative tax translates into a deferred tax asset).

```python
#Dados
receita = [14574.49, 7606.46, 8611.41, 9175.41, 8058.65, 8105.44, 11496.28, 9766.09, 10305.32, 14379.96, 10713.97, 15433.50]
despesa = [12051.82, 5695.07, 12319.20, 12089.72, 8658.57, 840.20, 3285.73, 5821.12, 6976.93, 16618.61, 10054.37, 3803.96]
```

```python
import numpy as np #importa a biblioteca Numpy
```

```python
aRec = np.array(receita) #Transforma a lista com a receita em um array e atribui a mesma a uma variável
aDesp = np.array(despesa) #Transforma a lista com as despesas em um array e atribui a mesma a uma variável
rCopia = aRec.copy() #Cria uma copia da array com a receita para que possa ser manipulada sem alterar a array original
dCopia = aDesp.copy() #Cria uma copia da array com as despesas para que possa ser manipulada sem alterar a array original
```

- ### profit for each month

```python
lLucro = [] #Cria uma lista vazia para armazenar valores
clLucro = []
somador = 0 #cria um somador com valor inicial de 0
for i in rCopia: #Cria um loop que itera sobre cada item armazenado na cópia da array com a receita
    lucro = i - dCopia[somador] #Calcula o lucro diminuindo a receita do mês pela despesa do mesmo mês
    clLucro.append(lucro)
    lucro = int(round(lucro/1000, 0)) #Tranforma o lucro do mês em unidade de 1000 e em um número inteiro
    lucro = str(lucro) + 'k' #Tranforma o lucro do mês em uma string e adiciona 'k', representando que o numero é uma unidade de 1000
    lLucro.append(lucro) #Adiciona o lucro a uma lista vazia criada anteriormente
    somador += 1 #Adiciona +1 ao somador, indicando que o calculo agora será com o mês seguinte  
print(lLucro) #Mostra uma lista com o lucro de cada mês em unidade de 1000
   
```

- ### profit after tax for each month (the tax rate is 30%)

```python
lLucroCImposto = [] #Cria uma lista vazia
clLucroCImposto = [] #cria uma lista vazia
for i in clLucro: #Cria um loop que irá iterar sobre cada item da lista lLucro
    lCImposto = round((i * 70) /100, 2) #Calcula os 70% de lucro, com os 30% de impostos retirados
    clLucroCImposto.append(lCImposto) #armazena o valor do lucro em formato de numero 
    lCImposto = ((int(round(lCImposto/1000, 0)))) #Transforma o valor para unidade de 1000
    lCImposto = str(lCImposto) + 'k' #Transforma o valor em uma string e adiciona 'k' a sua frente, representando que ele é unidade de 1000
    lLucroCImposto.append(lCImposto) #Adiciona o valor em formato de texto a lista vazia lLucroCImposto 
print(lLucroCImposto) #Mostra a lista com os lucros descontados os impostos 
```

- ### profit margin for each month - equals to profit after tax divided by revenue

```python
margemLucro = [] #Cria uma lista vazia
cmargemLucro = []
somador = 0 #Reseta o somador
for i in clLucroCImposto: #Cria um loop que itera sobre cada copia do item na lista do Lucro com o imposto aplicado 
    mLucro = int(round((i/rCopia[somador]) * 100, 0)) #Calcula a margem de lucro, transformando o valor em inteiro
    cmargemLucro.append(mLucro) #Adiciona o valor inteiro a uma lista criada anteriormente
    mLucro = str(mLucro) + '%' #transforma o valor em string e adiciona o símbolo % ao lado
    margemLucro.append(mLucro) ##Adiciona o valor inteiro a uma lista criada anteriormente
print(margemLucro) #Mostra a lista com a margem de lucro em %
```

- ### good months - where the profit after tax was greater than the mean for the year

```python
arrayLucroCImposto = np.array(clLucroCImposto) #Cria uma array com a lista que contém o inteiro do lucro depois de aplicado o imposto 
media = arrayLucroCImposto.mean() #Calcula a média da array e armazena na variável
lBonsMeses = [] #Cria uma lista vazia
for i in clLucroCImposto: #Cria um loop que itera sobre cada item da lista que contém o inteiro do lucro depois do imposto de cada mês
    if i > media: #Cria uma condição sob o item, com o fim de verificar se ele é maior do que a média do lucro depois de aplicado o imposto  
        valorBomMes = int(round(i/1000, 0)) #Transforma o número que atende a condição em uma unidade de 1000
        valorBomMes = str(valorBomMes) + 'k' #Transforma o valor em string e adiciona um "k" a frente
        lBonsMeses.append(valorBomMes) #Adiciona o valor a lista vazia criada anteriormente 
print(lBonsMeses) #Mostra a lista com os valores 
    
```

- ### bad months - where the profit after tax was less than the mean for the year

```python
lMesesRuins = [] #Cria uma lista vazia
for i in clLucroCImposto: #Cria um loop que itera sobre cada item da lista que contém o inteiro do lucro depois do imposto de cada mês
    if i < media: #Cria uma condição sob o item, com o fim de verificar se ele é menor do que a média do lucro depois de aplicado o imposto  
        valorMesRuim = int(round(i/1000, 0)) #Transforma o número que atende a condição em uma unidade de 1000
        valorMesRuim = str(valorMesRuim) + 'k' #Transforma o valor em string e adiciona um "k" a frente
        lMesesRuins.append(valorMesRuim) #Adiciona o valor a lista vazia criada anteriormente 
print(lMesesRuins) #Mostra a lista com os valores 
```

- ### the best month - where the profit after tax was max for the year
