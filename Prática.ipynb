#### Your program is to use the brute-force approach in order to find the Answer to Life, the Universe, and Everything. More precisely... rewrite small numbers from input to output. Stop processing input after reading in the number 42. All numbers at input are integers of one or two digits.

```python
#Criando lista vazia
output = []
x = 0
#Criando um loop que adiciona números digitado a lista e que para ao ser digitado o número 42
while x != 42:
    x = input('Insira um número de 0 a 99: ')
    try:
        x = int(x)
        if x > 99 or x < 0:
            print('Insira um número entre 0 e 99')
        else:
            x = int(x)
            if x == 42:
                break
            else:
                output.append(x)
    except:
        print('Formato inválido, digite um número inteiro')
#Imprimindo a lista
print(output)              
```

#### Peter wants to generate some prime numbers for his cryptosystem. Help him! Your task is to generate all prime numbers between two given numbers!



```python
#Criando uma lista vazia
lista = []
x = int(input('Digite o primeiro número: '))
y = int(input('Digite o segundo número: '))
# usando o comando for para percorrer um range entre dois número digitados pelo usuário
# e verificando se o número é primo
for i in range(x, y + 1):
    if i == 2 or i == 3 or i == 5 or i == 7:
        lista.append(i) 
    elif i == 1:
        continue
    elif i % 2 == 0 or i % 3 == 0 or i % 5 == 0 or i % 7 == 0:
        continue
    else:
        lista.append(i)
    
print(lista)
```

Reversed number is a number written in arabic numerals but the order of digits is reversed. The first digit becomes last and vice versa. Of course, the result is not unique because any particular number is a reversed form of several numbers (e.g. 21 could be 12, 120 or 1200 before reversing). #### According to Wikipedia, an arithmetic progression (AP) is a sequence of numbers such that the difference of any two successive members of the sequence is a constant. For instance, the sequence 3, 5, 7, 9, 11, 13, . . . is an arithmetic progression with common difference 2. For this problem, we will limit ourselves to arithmetic progression whose common difference is a non-zero integer. On the other hand, a geometric progression (GP) is a sequence of numbers where each term after the first is found by multiplying the previous one by a fixed non-zero number called the common ratio. For example, the sequence 2, 6, 18, 54, . . . is a geometric progression with common ratio 3. For this problem, we will limit ourselves to geometric progression whose common ratio is a non-zero integer. Given three successive members of a sequence, you need to determine the type of the progression and the next successive member.

```python
#criando uma string para usar como parâmetro para validar a entrada de dados 
validator = False
#validando a entrada de dados do número a ser inserido
while validator == False:
    num1 = input('Insira o primeiro número: ')
    try:
        num1 = int(num1)
        if num1 < 0:
            print('Insira um inteiro maior do que zero')
        else:
            validator = True
    except: 
        print('Formato inválido, insira um número inteiro')
#Recolocando o parâmetro de validação ao valor original para ser usado novamente
validator = False
#validando a entrada de dados do número a ser inserido
while validator == False:
    num2 = input('Insira o segundo número: ')
    try:
        num2 = int(num2)
        if num2 < 0:
            print('Insira um inteiro maior do que zero')
        else:
            validator = True
    except: 
        print('Formato inválido, insira um número inteiro')
#Recolocando o parâmetro de validação ao valor original para ser usado novamente
validator = False
#validando a entrada de dados do número a ser inserido
while validator == False:
    num3 = input('Insira o terceiro número: ')
    try:
        num3 = int(num3)
        if num3 < 0:
            print('Insira um inteiro maior do que zero')
        else:
            validator = True
    except: 
        print('Formato inválido, insira um número inteiro')
#Criando variáveis para armazenar valores a serem testados
valuePA = num2 - num1
value1PA = num3 - num2
valuePG = num2 / num1
value1PG = num3 / num2
#Testando os valores armazenados anteriormente para tentar identificar se os números digitados 
# representa progressão aritmética, progressiva ou nenhuma
if valuePA == value1PA:
    print('AP', int(num3 + valuePA))
elif valuePG == value1PG:
        print('GP', int(num3 * valuePG))
else:
    print('Nenhuma progressão identificada')
```

#### Problem statement is very easy . On a positive integer, you can perform any one of the following 3 steps.

#### 1.)  Subtract 1 from it. ( n = n - 1 ) 

#### 2.)  If its divisible by 2, divide by 2. ( if n % 2 == 0 , then n = n / 2  )

#### 3.)  If its divisible by 3, divide by 3. ( if n % 3 == 0 , then n = n / 3  )

#### Given a positive integer  n and you task is find the minimum number of steps that takes n to one .

 

```python
#criando uma função para verificar a quantidade de oeprações necessárias para reduzir o número  informado
# ao menor número possível dividindo por 3 ou  2
def passos(n):
    passo = 0
    while n > 1:
        if n % 3 == 0:
            n = n/3
        elif n % 2 == 0:
            n = n/2
        else:
            n -= 1
        passo += 1
    print('Parar chegar no menor número, foram seguidos %r passos.' %(passo))
    return
    
    
#-----------------------------------------------------------------------------------------    
#criando uma string para usar como parâmetro para validar a entrada de dados 
valid = False
#validando a entrada de dados do número a ser inserido
while valid == False:
    num = input('Digite o número: ')
    try:
        num = int(num)
        if num < 0:
            print('Insira um número inteiro maior do que 0')
        else: 
            valid = True
    except:
        print('Formato inválido, insira um número inteiro maior que 0')
#Chamando a função criada no início e passando um parâmetro        
passos(num)        
```

#### The purpose of this problem is to verify whether the method you are using to read input data is sufficiently fast to handle problems branded with the enormous Input/Output warning. You are expected to be able to process at least 2.5MB of input data per second at runtime.

```python
#Criando listas vazias
finallist = []
valuelist = []
#criando uma string para usar como parâmetro para validar a entrada de dados 
validator = False
#validando a quantidade de números que será inserido
while validator == False:
    n = input('Insira quantos números você irá inserir: ')
    try:
        n = int(n)
        if n <= 0:
            print('insira uma quantidade de vezes maior que 0')
        else:
            validator = True
    except:
        print('Formato inválido, insira um número interiro e positivo')
#Recolocando o parâmetro de validação ao valor original para ser usado novamente        
validator = False     
#Validando os números que serão divididos 
while n > 0:
    while validator == False:
        number = input('insira um número: ')
        try:
            number = int(number)
            if number <= 0:
                print('insira um número inteiro maior que 0')
            else:
                validator = True
        except:
            print('Formato inválido, insira um número inteiro')
    #Adicionando os números digitados em uma lista e reduzindo a quantidade de números a ser inseridos
    valuelist.append(number)
    n -= 1
    #Recolocando o parâmetro de validação ao valor original para a validação dos números no loop seja refeita  
    validator = False
#Recolocando o parâmetro de validação ao valor original para ser usado novamente     
validator = False    
#validando o número que será usado como divisor 
while validator == False:
    numdivisor = input('Digite o divisor: ')
    try:
        numdivisor = int(numdivisor)
        if numdivisor <= 0:
            print('insira um divisor maior que 0')
        else:
            validator = True
    except:
        print('Formato inválido, insira um número interiro e positivo')
        
#Percorrendo a lista de números digitados, verificando se eles são divisíveis pelo divisor inserido e adicionando 
#os divísiveis em uma nova lista, imprimindo o comprimento da mesma no final 
for i in valuelist:
    if i % numdivisor == 0:
        finallist.append(i)
    else:
        continue
print('Há %r números divisíveis por %r' %(len(finallist), numdivisor))
```

#### The famous knapsack problem. You are packing for a vacation on the sea side and you are going to carry only one bag with capacity S (1 <= S <= 2000). You also have N (1<= N <= 2000) items that you might want to take with you to the sea side. Unfortunately you can not fit all of them in the knapsack so you will have to choose. For each item you are given its size and its value. You want to maximize the total value of all the items you are going to bring. What is this maximum total value?



```python
#Criando uma lista vazia
listavalues = []
#criando uma string para usar como parâmetro para validar a entrada de dados 
validade = False
#validando a entrada de dados para o inteiro "item"
while validade == False:
    item = input('Digite quantos itens você irá levar: ')
    try:
        item = int(item)
        if item < 0:
            print('Insira um número de itens maior do que 0')
        else:
            validade = True
    except:
        print('Formato inválido, insira um valor inteiro maior do que 0')
#Recolocando o parâmetro de validação ao valor original para ser usado novamente
validade = False        
#Criando um loop para registrar o valor final do item baseado no seu peso e seu tamanho 
while item > 0:
    #Validando a entrada de dados para o inteiro "tamanhoItem"
    while validade == False:
        tamanhoItem = input('Digite o tamanho do item: ')
        try:
            tamanhoItem = int(tamanhoItem)
            if tamanhoItem < 1:
                print('Insira um tamanho superior a 0')
            else:
                validade = True
        except:
            print('Insira um formato válido')
    #Recolocando o parâmetro de validação ao valor original para ser usado novamente      
    validade = False
    
    while validade == False:
        pesoItem = input('Digite o peso do item: ')
        try:
            pesoItem = int(pesoItem)
            if pesoItem < 0:
                print('Insira um numero inteiro')
            else:
                validade = True
        except:
            print('Insira um formato válido')
    #Recolocando o parâmetro de validação ao valor original para ser usado novamente no loop
    validade = False    
    item -= 1
    #Obtendo o valor final do item e adicionando o mesmo a uma lista
    valorfinal = tamanhoItem + pesoItem
    listavalues.append(valorfinal)
#Ordenando a lista em ordem numérica
listavalues.sort()
#Registrando na variável value a posição média dos valores, onde está a melhor combinação 
value = int(len(listavalues)/2)
#Imprimindo a melhor combinação baseada na posição dos itens na lista
print('A melhor combinação de itens é: ', listavalues[value - 1] + listavalues[value])
```
